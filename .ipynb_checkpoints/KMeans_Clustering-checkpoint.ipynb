```python
#import all relevant modules
import pandas as pd

from matplotlib import pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
#read in the correct dataset
df = pd.read_csv("crash_data_one_hot_encoded_MLP.csv")
df.head()
```

```python
#read in the correct dataset
df2 = pd.read_csv("crash_data_one_hot_encoded_MLP.csv")
df.head()
```

```python
#creates scale for numeric data (0-1)
#df.drop('TRAFFIC_CTRL', axis='columns', inplace=True)
df.drop('LG_REGION_DESC_0', axis='columns', inplace=True)
df.drop('DIRN_ROLE1_DESC_0', axis='columns', inplace=True)
df.drop('DIRN_ROLE1_DESC_Unknown', axis='columns', inplace=True)
df.drop('JUNCTION_TYPE_Unknown', axis='columns', inplace=True)
df.drop('FLAT_HILL_Unknown', axis='columns', inplace=True)
df.drop('ROAD_CURVATURE_Unknown', axis='columns', inplace=True)
df.drop('ROAD_MARKINGS_Unknown', axis='columns', inplace=True)
df.drop('ROAD_SURFACE_Unknown', axis='columns', inplace=True)
df.drop('ROAD_WET_Unknown', axis='columns', inplace=True)
df.drop('LIGHT_Unknown', axis='columns', inplace=True)
df.drop('STREET_LIGHT_Unknown', axis='columns', inplace=True)
df.drop('WEATHER_A_Unknown', axis='columns', inplace=True)
df.to_csv('final_dataset_without_clusters.csv', index=False)
df_with_CS_code = df
df.drop('CRASH_SEV_CODE', axis='columns', inplace=True)
scaler = MinMaxScaler()
scaler.fit(df)
scaled_df = scaler.transform(df)
#need to change array to df before i can do this
#scaled_df.to_csv('scaled_crash_data.csv', index=False)
```

```python
#find all columns to consider
count = 0
for col in df.columns: 
    print("'" + col + "'" + ", " + str(count)) 
    count= count+1
```

```python
#create KMeans model
#kmeans = KMeans(n_clusters=2)
kmeans_scaled = KMeans(n_clusters=7)
#y_predicted = kmeans.fit_predict(df)
y_scaled_predicted = kmeans_scaled.fit_predict(scaled_df)
#print(y_predicted)
print(y_scaled_predicted)
```

```python
#find cluster centres 
#df.drop('CLUSTER', axis='columns', inplace=True)

#print(kmeans.cluster_centers_)
print(kmeans_scaled.cluster_centers_)
```

```python
#add clusters as a column on dataframe
#df['CLUSTER']=y_predicted
df_with_CS_code['CLUSTER']=y_scaled_predicted
df_with_CS_code
```

```python
#export as csv with cluster data
df_with_CS_code.to_csv('final_dataset.csv', index=False)
```

```python
#Find the sum of squared error for each number of clusters
sse_scaled = []
sse = []
k_rng = range(1,10)
for k in k_rng:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    kmeans_scaled = KMeans(n_clusters=k)
    kmeans_scaled.fit(scaled_df)
    sse.append(kmeans.inertia_)
    sse_scaled.append(kmeans_scaled.inertia_)
print("finished")
```

```python
#Plot line graph that shows the sum of squared error for each number of k values
# appropriate number of clusters is at the elbow of the plot, the non-scaled clusters looks like 2
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng, sse)
```

```python
#Plot line graph that shows the sum of squared error for each number of k values
# appropriate number of clusters is at the elbow of the plot, the scaled clusters looks like 5
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng, sse_scaled)
```

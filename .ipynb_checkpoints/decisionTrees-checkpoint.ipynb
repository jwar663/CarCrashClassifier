```python
import multiprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import time

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import f1_score

from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold

from matplotlib.legend_handler import HandlerLine2D

#feature importance graphing function
def f_importances(importances, names):
    indices = np.argsort(importances)[::-1]
    print(indices)
    plt.figure()
    plt.subplots(figsize=(20,20))
    plt.title("Feature Importance")
    plt.bar(range(X.shape[1]), importances[indices])
    plt.xticks(range(X.shape[1]), names[indices], rotation=90)
    plt.show()



df = pd.read_csv("crash_data_one_hot_encoded_MLP.csv")
df = sklearn.utils.shuffle(df)
df.fillna(0,inplace=True)
encode = LabelEncoder()

features = ['CRASH_YEAR', 'NUM_LANES', 'SPD_LIM', 'MULTI_VEH_CODE','HOLIDAY_CODE',
            'LG_REGION_DESC_CODE', 'JUNCTION_TYPE_CODE', 'DIRN_ROLE1_DESC_CODE','INTSN_MIDBLOCK_CODE',
            'FLAT_HILL_CODE','ROAD_CURVATURE_CODE','ROAD_MARKINGS_CODE','ROAD_SURFACE_CODE','ROAD_WET_CODE',
            'URBAN_CODE','LIGHT_CODE','STREET_LIGHT_CODE','WEATHER_A_CODE']

df_non_injury = df[df.CRASH_SEV_CODE==2]  # 480452 samples
df_minor = df[df.CRASH_SEV_CODE==1]       # 150834 samples
df_serious = df[df.CRASH_SEV_CODE==3]    # 37347  samples
df_fatal = df[df.CRASH_SEV_CODE==0]       # 6178   samples

#upsampling all classes to match largest class (non_injury)
df_minor_upsampled = resample(df_minor, replace=True,n_samples=480452,random_state=40)
df_serious_upsampled = resample(df_serious, replace=True,n_samples=480452,random_state=40)
df_fatal_upsampled = resample(df_fatal, replace=True,n_samples=480452,random_state=40)

#downsampling all classes to match smallest class (fatal)
df_minor_downsampled= resample(df_minor, replace=False,n_samples=6178,random_state=40)
df_serious_downsampled = resample(df_serious, replace=False,n_samples=6178,random_state=40)
df_non_injury_downsampled = resample(df_non_injury, replace=False,n_samples=6178,random_state=40)


df_upsampled = pd.concat([df_non_injury, df_minor_upsampled, df_serious_upsampled, df_fatal_upsampled])
df_upsampled.CRASH_SEV_CODE.value_counts()

df_downsampled = pd.concat([df_fatal, df_minor_downsampled, df_serious_downsampled, df_non_injury_downsampled])
df_downsampled.CRASH_SEV_CODE.value_counts()
```

```python
X = df_upsampled.drop(["CRASH_SEV_CODE","CRASH_YEAR"], axis=1).values
y = df_upsampled["CRASH_SEV_CODE"].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=40)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

clf = DecisionTreeClassifier(random_state = 30)

print(X_train.shape); print(X_test.shape)
```

```python

start = time.time()
clf.fit(X_train,y_train)
stop = time.time()

predict_train = clf.predict(X_train)
predict_test = clf.predict(X_test)
print(f"\n Training time: {stop - start}s")
print("\n training data\n")
print(confusion_matrix(y_train,predict_train))
print("\n")
print(classification_report(y_train,predict_train,labels=pd.unique(predict_train)))
print("\n---------------------------------------------------------------------------------------\n")
print("\n test data\n")
print(confusion_matrix(y_test,predict_test))
print("\n")
print(classification_report(y_test,predict_test,labels=pd.unique(predict_test)))
```

```python
#export decision tree model for testing
from joblib import dump, load
dump(clf, 'decision_tree_classification_model.joblib')
```

```python
scores = []
kf = KFold() #defualt 5 folds
for train_index, test_index in kf.split(X):
    print("Train Index: ", train_index, "\n")
    print("Test Index: ", test_index)

    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
```

```python
max_depths = np.linspace(1, 100, 100, endpoint=True)
train_results = []
test_results = []
for max_depth in max_depths:
   dt = DecisionTreeClassifier(max_depth=max_depth)
   dt.fit(X_train, y_train)
    
   predict_train = dt.predict(X_train)
   score = f1_score(y_train,predict_train,labels=pd.unique(predict_train),average='weighted')
   train_results.append(score)

   predict_test = dt.predict(X_test)
   score1 = f1_score(y_test,predict_test,labels=pd.unique(predict_test),average='weighted')
   test_results.append(score1)
    

line1, = plt.plot(max_depths, train_results, 'b', label="Train AUC")
line2, = plt.plot(max_depths, test_results, 'r', label="Test AUC")
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('f1_score')
plt.xlabel('Tree depth')
plt.show()
```

```python
min_splits = range(2,10)
train_results1 = []
test_results1 = []
for splits in min_splits:
   dt = DecisionTreeClassifier(min_samples_split = splits)
   dt.fit(X_train, y_train)
    
   predict_train = dt.predict(X_train)
   score = f1_score(y_train,predict_train,labels=pd.unique(predict_train),average='weighted')
   train_results1.append(score)

   predict_test = dt.predict(X_test)
   score1 = f1_score(y_test,predict_test,labels=pd.unique(predict_test),average='weighted')
   test_results1.append(score1)
    
line1, = plt.plot(min_splits, train_results1, 'b', label="Train score")
line2, = plt.plot(min_splits, test_results1, 'r', label="Test score")
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('f1_score')
plt.xlabel('min samples split')
plt.show()
```

```python
min_samples_leafs  = range(1,10)
train_results2 = []
test_results2 = []
for min_samples in min_samples_leafs:
   dt = DecisionTreeClassifier(min_samples_leaf = min_samples)
   dt.fit(X_train, y_train)
    
   predict_train = dt.predict(X_train)
   score = f1_score(y_train,predict_train,labels=pd.unique(predict_train),average='weighted')
   train_results2.append(score)

   predict_test = dt.predict(X_test)
   score1 = f1_score(y_test,predict_test,labels=pd.unique(predict_test),average='weighted')
   test_results2.append(score1)
    
line1, = plt.plot(min_samples_leafs, train_results2, 'b', label="Train score")
line2, = plt.plot(min_samples_leafs, test_results2, 'r', label="Test score")
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('f1_score')
plt.xlabel('min samples leaf')
plt.show()
```

```python
n_estimators = [1, 2, 4, 8, 16, 32, 64, 100, 200]
train_results4 = []
test_results4 = []
training_time4 = [] # in mins
for estimator in n_estimators:
   rf = RandomForestClassifier(n_estimators=estimator,n_jobs=-1, verbose = True)

   start = time.time()
   rf.fit(X_train, y_train)
   stop = time.time()
    
   training_time4.append((stop - start)/60)
    
    
   predict_train = rf.predict(X_train)
   score = f1_score(y_train,predict_train,labels=pd.unique(predict_train),average='weighted')
   train_results4.append(score)

   predict_test = rf.predict(X_test)
   score1 = f1_score(y_test,predict_test,labels=pd.unique(predict_test),average='weighted')
   test_results4.append(score1)
    
from matplotlib.legend_handler import HandlerLine2D
line1, = plt.plot(n_estimators, train_results4, 'b', label="Train score")
line2, = plt.plot(n_estimators, test_results4, 'r', label="Test score")
plt.legend(handler_map={line1: HandlerLine2D(numpoints=2)})
plt.ylabel('f1_score')
plt.xlabel('min samples leaf')
plt.show()

time_line, = plt.plot(n_estimators, training_time4, 'r', label="time taken")
plt.legend(handler_map={time_line: HandlerLine2D(numpoints=1)})
plt.ylabel('time (mins)')
plt.xlabel('num trees')
plt.show()
```
